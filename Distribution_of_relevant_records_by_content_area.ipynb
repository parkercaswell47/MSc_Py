```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('NbS_Assessment_2.1.csv')

x = data["Aspect"]
y = data["n"]
# Function to add value labels
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(y[i] + 0.5, i, f"n = {y[i]}", va='center', fontsize=11)  # slightly larger text

# Set figure size using golden ratio
plt.figure(figsize=(10, 6.18))

# Create horizontal bar chart
plt.barh(x, y, color="#2B3D50")

# Add value labels
add_labels(x, y)

# Titles and axes labels with larger font sizes
plt.title("Distribution of Relevant Records by Content Area", fontsize=14)
plt.xlabel("Number of Records", fontsize=12)

# Format x-axis
plt.xticks(range(0, max(y) + 5, 5), fontsize=11)
plt.yticks(fontsize=11)
plt.xlim(0, max(y) + 3)

# Show plot
plt.show()
```
